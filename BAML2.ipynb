```python
# EDA & Cleaning Prep for Colab

import pandas as pd

# Load data
main_df = pd.read_csv('diabetic_data.csv')

# Replace '?' with NaN
main_df.replace('?', pd.NA, inplace=True)

# Drop columns with too many missing values or irrelevant identifiers
cols_to_drop = ['weight', 'payer_code', 'medical_specialty', 'encounter_id', 'patient_nbr']
main_df.drop(columns=cols_to_drop, inplace=True)

# Optional: Convert age brackets to numeric midpoint
main_df['age'] = main_df['age'].str.extract(r'(\d+)', expand=False).astype(float) + 5

# Encode 'readmitted' into binary classification
main_df['readmitted'] = main_df['readmitted'].replace({'NO': 0, '>30': 0, '<30': 1})

# Drop duplicates, if any
main_df.drop_duplicates(inplace=True)

# Save cleaned version
main_df.to_csv('cleaned_diabetic_data.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Read the cleaned data
df = pd.read_csv('cleaned_diabetic_data.csv')

# Class distribution
sns.countplot(x='readmitted', data=df)
plt.title('Readmission Distribution')
plt.show()

# Heatmap of correlations (numeric only)
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True)
plt.title('Feature Correlation')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('diabetic_data.csv')

# Data Cleaning
df.replace('?', pd.NA, inplace=True)
df['age'] = df['age'].str.extract(r'(\d+)', expand=False).astype(float) + 5
df['readmitted_binary'] = df['readmitted'].replace({'NO': 0, '>30': 0, '<30': 1}).astype(int)
df.drop(columns=['encounter_id', 'patient_nbr', 'weight', 'payer_code', 'medical_specialty'], inplace=True)
df['gender'] = df['gender'].astype('category')
df['race'] = df['race'].astype('category')


```

```python
# Plot 1: Readmission Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='readmitted_binary', data=df)
plt.title('Readmission within 30 Days')
plt.xticks([0, 1], ['No', 'Yes'])
plt.xlabel('Readmitted')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

```

```python
# Plot 2: Age Distribution by Readmission
plt.figure(figsize=(8, 5))
sns.boxplot(x='readmitted_binary', y='age', data=df)
plt.xticks([0, 1], ['No', 'Yes'])
plt.title('Age Distribution by Readmission')
plt.xlabel('Readmitted')
plt.ylabel('Age (midpoint of range)')
plt.tight_layout()
plt.show()

```

```python
# Plot 3: Race vs Readmission
plt.figure(figsize=(10, 5))
sns.countplot(x='race', hue='readmitted_binary', data=df)
plt.title('Readmission by Race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.legend(title='Readmitted', labels=['No', 'Yes'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
