```python
correlation_with_target = df.corr()['X65'].sort_values(ascending=False)

# Display top correlations
print("Top 10 positively correlated features with bankruptcy:")
print(correlation_with_target[1:11])
print("\nTop 10 negatively correlated features with bankruptcy:")
print(correlation_with_target[-10:])

# Plot top correlations
plt.figure(figsize=(12, 10))
top_features = pd.concat([correlation_with_target[1:11], correlation_with_target[-10:]])
colors = ['red' if c > 0 else 'green' for c in top_features.values]
plt.barh(top_features.index, top_features.values, color=colors)
plt.axvline(x=0, color='black', linestyle='-', alpha=0.3)
plt.title('Features Most Correlated with Bankruptcy')
plt.xlabel('Correlation Coefficient')
plt.tight_layout()
plt.show()
```

```python

'''
Company size is imp - The logarithm of total assets (X29) has the strongest neg corr with bankruptcy, indicating that smaller companies face way higher bankruptcy risk regardless of other financial factors.

Debt burden increases failure risk - Total liabilities to total assets (X2) shows the strongest pos corr with bankruptcy, confirming that higher debt increases bankruptcy prob.

Short-term debt is particularly dangerous - Short-term liabilities to total assets (X51) has the second highest pos corr corr--immediate debt reqa create more pressure than longterm debt.

Working capital is imp for survival - Both working capital metrics (X3, X55) show strong negcorr with bankruptcy, so imp to maintain adequate to be ready cash.

Companies that retain more earnings (X6) show lower bankruptcy risk (-0.049), building financial reserves is imp.

Correlations are strangely low overall - bankruptcy prediction requires multivariate models rather than relying on individual financial ratios. So data is good and no feature dominance.
'''
```

```python
import numpy as np
import scipy.cluster.hierarchy as hierarchy
from scipy.stats import spearmanr

# Compute Spearman correlation matrix
corr = spearmanr(X_scaled).correlation

# Convert correlation matrix to a valid distance matrix
distance_matrix = 1 - np.abs(corr)

# Ensure strict symmetry (fix numerical precision issues)
distance_matrix = (distance_matrix + distance_matrix.T) / 2

# Ensure zero diagonal
np.fill_diagonal(distance_matrix, 0)

# Convert to condensed form for clustering
corr_condensed = hierarchy.distance.squareform(distance_matrix)

# Perform hierarchical clustering
z = hierarchy.linkage(corr_condensed, method='average')
```

```python

plt.figure(figsize=(16, 10))
dendro = hierarchy.dendrogram(z, labels=X.columns, leaf_rotation=90)
plt.title('Feature Clustering by Spearman Correlation')
plt.xlabel('Features')
plt.ylabel('Distance (1 - |Correlation|)')
plt.axhline(y=0.5, color='r', linestyle='--', alpha=0.5)  # Suggested cluster cutoff
plt.tight_layout()
plt.show()
```

```python
   from sklearn.preprocessing import PowerTransformer, RobustScaler
   from sklearn.feature_selection import SelectFromModel
   from sklearn.decomposition import PCA
   from sklearn.ensemble import RandomForestClassifier
```

```python
def preprocess_bankruptcy_data(df, n_components=10, n_features_to_select=15, random_state=42):


   X = df.drop('X65', axis=1)
   y = df['X65']
   X_processed = X.copy()
   
   positive_skew_features = ['X20', 'X43', 'X44', 'X58', 'X62']
   negative_skew_features = ['X16', 'X45', 'X49', 'X56']
   
   # Handle zeros in specific features where zeros have meaning
   meaningful_zero_features = ['X22', 'X42', 'X59']
   for feature in meaningful_zero_features:
       if feature in X_processed.columns:
           # Create binary indicator for zero values
           X_processed[f'{feature}_is_zero'] = (X_processed[feature] == 0).astype(int)
           # Replace zeros with a small value for log transforms
           X_processed[feature] = X_processed[feature].replace(0, 1e-6)
   
   # Apply log transformation to positive skewed features
   for feature in positive_skew_features:
       if feature in X_processed.columns:
           min_val = X_processed[feature].min()
           if min_val <= 0:
               X_processed[feature] = X_processed[feature] - min_val + 1  # Shift to positive
           X_processed[feature] = np.log1p(X_processed[feature])
   
   # Apply power transformation to negative skewed features
   pt = PowerTransformer(method='yeo-johnson')
   if negative_skew_features:
       negative_skew_cols = [col for col in negative_skew_features if col in X_processed.columns]
       if negative_skew_cols:
           X_processed[negative_skew_cols] = pt.fit_transform(X_processed[negative_skew_cols])
   
   # Apply robust scaling to handle outliers
   scaler = RobustScaler()
   X_scaled = scaler.fit_transform(X_processed)
   X_scaled_df = pd.DataFrame(X_scaled, columns=X_processed.columns)
   
   # Feature selection using Random Forest importance
   selector = SelectFromModel(
       RandomForestClassifier(n_estimators=100, random_state=random_state),
       max_features=n_features_to_select
   )
   selector.fit(X_scaled, y)
   selected_mask = selector.get_support()
   selected_features = X_processed.columns[selected_mask].tolist()
   X_selected = X_scaled_df.iloc[:, selected_mask]
   
   # Apply PCA to the selected features
   pca = PCA(n_components=min(n_components, len(selected_features)), random_state=random_state)
   X_pca = pca.fit_transform(X_selected)
   
   # Create a DataFrame with the principal components
   pca_columns = [f'PC{i+1}' for i in range(X_pca.shape[1])]
   X_transformed = pd.DataFrame(X_pca, columns=pca_columns)
   
   # Calculate and store explained variance
   explained_variance = pca.explained_variance_ratio_
   cumulative_variance = np.cumsum(explained_variance)
   
   print(f"Selected features: {selected_features}")
   print(f"Number of components: {X_pca.shape[1]}")
   print(f"Explained variance: {cumulative_variance[-1]:.2f}")
   
   return X_transformed, y, pca_columns, {
       'selected_features': selected_features,
       'pca': pca,
       'scaler': scaler,
       'positive_skew_features': positive_skew_features,
       'negative_skew_features': negative_skew_features,
       'meaningful_zero_features': meaningful_zero_features
   }



X_transformed, y, feature_names, preprocessing_info = preprocess_bankruptcy_data(df)
```
